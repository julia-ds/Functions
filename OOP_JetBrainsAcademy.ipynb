Write a function that allows a person to buy coffee if there is enough resources left, check the remaining quantities of ingridients, take money or fill the machine up with desired amounts of water, milk, beans, cups.
Your coffee machine should have the following initial resources: 400 ml of water, 540 ml of milk, 120 g of coffee beans, 9 disposable cups, $550 in cash.

```python
class Coffee:
    def __init__(self, water, milk, beans, cup, cost):
        self.water = water
        self.milk = milk
        self.beans = beans
        self.cup = cup
        self.cost = cost
        
class CoffeeType:
    espresso = Coffee(250, 0, 16, 1, 4)
    latte = Coffee(350, 75, 20, 1, 7)
    cappuccino = Coffee(200, 100, 12, 1, 6)
    
class Components:
    def __init__(self):
        self.water = 400
        self.milk = 540
        self.beans = 120
        self.cup = 9
        self.money = 550
        
    def coffee_preparation(self, coffee):
        water = self.water - coffee.water
        milk = self.milk - coffee.milk
        beans = self.beans - coffee.beans
        cup = self.cup - coffee.cup
        if water < 0:
            print("Sorry, not enough water!")
        elif milk < 0:
            print("Sorry, not enough milk!")
        elif beans < 0:
            print("Sorry, not enough beans!")
        elif cup < 0:
            print("Sorry, not enough disposable cups!")
        else:
            print("I have enough resources, making you a coffee!")
            self.water = water
            self.milk = milk
            self.beans = beans
            self.cup = cup
            self.money += coffee.cost
        
    def make_coffee(self):
        answer = input("What do you want to buy? 1 - espresso, 2 - latte, 3 - cappuccino, back - to main menu:")
        if answer == "1":
            self.coffee_preparation(CoffeeType.espresso)
        elif answer == "2":
            self.coffee_preparation(CoffeeType.latte)
        elif answer == "3":
            self.coffee_preparation(CoffeeType.cappuccino)
        
    def remaining(self):
        print("The coffee machine has:")
        print("%d of water" %self.water)
        print("%d of milk" %self.milk)
        print("%d of coffee beans" %self.beans)
        print("%d of disposable cups" %self.cup)
        print("%d of money" %self.money)
    
    def fill(self):
        w_ml = int(input("Write how many ml of water do you want to add:"))
        self.water += w_ml
        m_ml = int(input("Write how many ml of milk do you want to add:"))
        self.milk += m_ml
        b = int(input("Write how many grams of coffee beans do you want to add:"))
        self.beans += b
        c = int(input("Write how many disposable cups of coffee do you want to add:"))
        self.cup += c
        
    def take(self):
        print("I gave you $" + str(self.money))
        self.money = 0
    
class CoffeeMachine:
    def __init__(self):
        self.coffee_type = CoffeeType()
        self.components = Components()
        
    def running(self):
        while True:
            print("Write action (buy, fill, take, remaining, exit):")
            action = input()
            if action == 'buy':
                self.components.make_coffee()
            elif action == 'fill':
                self.components.fill()
            elif action == 'take':
                self.components.take()
            elif action == 'remaining':
                self.components.remaining()
            elif action == 'exit':
                break
```

```python
machine = CoffeeMachine()
CoffeeMachine.running(machine)
```
